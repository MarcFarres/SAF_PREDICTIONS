```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
errors_linear = np.load("errors_linear.npy")
errors_linear_nowinter = np.load("errors_linear_nowinter.npy")
errors_ML = np.load("errors_ML.npy")[:100]
errors_ML_nowinter = np.load("errors_ML_nowinter.npy")[:100]
```

```python
plt.plot(errors_linear, color="blue", label="Linear model")
plt.plot(errors_ML, color="red", label="ML-based model")
plt.title("MAE per step in the future")
plt.ylabel("MAE")
plt.xlabel("Number of steps in the future")
plt.xlim(left=0)
plt.ylim(bottom=0)
plt.legend(title="Errors", loc="upper left")
plt.yticks([0.0, 0.005, 0.01, 0.015, 0.02, 0.025])
plt.savefig("errors_winter.eps", format="eps")
plt.show()
```

```python
plt.plot(errors_linear_nowinter, color="blue", label="Linear model")
plt.plot(errors_ML_nowinter, color="red", label="ML-based model")
plt.title("MAE per step in the future")
plt.ylabel("MAE")
plt.xlabel("Number of steps in the future")
plt.xlim(left=0)
plt.ylim(bottom=0)
plt.legend(title="Errors", loc="upper left")
plt.yticks([0.0, 0.005, 0.01, 0.015, 0.02, 0.025])
plt.savefig("errors_nowinter.eps", format="eps")
plt.show()
```

```python
plt.plot(errors_linear)
plt.title("MAE per step in the future")
plt.ylabel("MAE")
plt.xlabel("Number of steps in the future")
plt.xlim(left=0)
plt.ylim(bottom=0)
plt.savefig("linear_error.eps", format="eps")
plt.show()
idx = np.argmax(errors_linear)
print(f"Highest error is : {errors_linear[idx]} on {idx} steps")
```

```python
plt.plot(errors_ML)
plt.title("MAE per step in the future")
plt.ylabel("MAE")
plt.xlabel("Number of steps in the future")
plt.xlim(left=0)
plt.ylim(bottom=0)
plt.savefig("ML_error.eps", format="eps")
plt.show()
idx = np.argmax(errors_ML)
print(f"Highest error is : {errors_ML[idx]} on {idx} steps")
```

```python
plt.plot(errors_ML_nowinter)
plt.title("MAE per step in the future")
plt.ylabel("MAE")
plt.xlabel("Number of steps in the future")
plt.xlim(left=0)
plt.ylim(bottom=0)
plt.savefig("Nowinter.eps", format="eps")
plt.show()
idx = np.argmax(errors_ML_nowinter)
print(f"Highest error is : {errors_ML_nowinter[idx]} on {idx} steps")
```
